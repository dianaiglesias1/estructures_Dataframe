#### Resumeix estadísticament les columnes d'interès

```python
#Summarize statistically the columns. We'll use describe() function

subset_airlines.describe()
```

```python
#let's take a look in detail one of the columns

subset_airlines.ArrDelay.describe().round(1)
```

As we know the values in this column are in minutes we can say that the average delay among all observations is 42 mins and that most of the flights have a delay time larger than the average

```python
#Column 'month'
#Now, I want to know how many flights have operated each month

subset_airlines['Month'].value_counts()
```

So basically March, June and December are the months with more flights. It makes sense...people intensify travels for Easter, Summertime, and Christmas.

```python
#Even though We got the data, I need it more clear and ordered by month. So I will: create a dataframe, change numbers
#by name of months and change the name index of column 2 to "Number of flights"

by_month=airlines['Month'].value_counts().to_frame().sort_index()
by_month.index=['January','February','March','April','May','June','July','August','September','October','November','December']
by_month.rename(columns={
        'Month': 'Number of flights', 
    }, inplace=True)

by_month
```

```python
#I want all this information on a graph

plt.rcParams['figure.figsize'] = (15, 5)  #set graph config (size)
plt.rcParams['font.family'] = 'sans-serif' #set graph font
by_month['Number of flights'].plot(kind='bar', title='Number of flights by month'); #plot graph and set title
plt.xlabel("Months") #name x axis
plt.ylabel("Number of flights") #name y axis
```

```python
#I would like to add a third column on the dataframe with the percentages by month

by_month['%'] = ((by_month['Number of flights'] / by_month['Number of flights'].sum())*100).round(2)
by_month.sort_values(by=['%'],ascending=False)
```

Then, specifically we can say that December, March and June have the highest % of flights during the year.

Great! But what if we want to know **which day of the week is the most busy in terms of flights?**

```python
#Column 'DayofWeek'

by_day_week=airlines['DayOfWeek'].value_counts().to_frame().sort_index()
by_day_week.index=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
by_day_week.rename(columns={
    'DayOfWeek':'Flights per week',
  }, inplace=True)
by_day_week
```

```python
#now, with %

by_day_week['%'] = ((by_day_week['Flights per week'] / by_day_week['Flights per week'].sum())*100).round(2)
by_day_week
```

```python
#I want it to be sorted by % in descending value. As we are already working with a dataframe we'll use 
#sort_values as before but adding other parameters to the function

by_day_week.sort_values(by=['%'],ascending=False)
```

So Friday and Monday are the busiest days of the week...people going back to home after business week and people visting other places for the weekends...

```python
#Column Cancelled

#Now, I want to know how many cancelled flights have been during the year

total_cancelled=sum(airlines.Cancelled)
print('There has been',total_cancelled,'flights cancelled')
```

```python
#Nice, but it means nothing to me...what is the % of cancelled flights over total flights?

total_flights_year=sum(airlines.Month)
percentage_cancelled_flights=round((total_cancelled/total_flights_year)*100,3)
print("The % of cancelled flights over total flights is", percentage_cancelled_flights,'%')
```

```python
#Column UniqueCarrier

#I need to know hoy many flights has operated each airline during the year

by_airline=airlines.UniqueCarrier.value_counts().to_frame()
by_airline.rename(columns={
    'UniqueCarrier':'AnnualFlights'
    },inplace=True)
by_airline['% of Total Annual Flights'] = ((by_airline['AnnualFlights'] / by_airline['AnnualFlights'].sum())*100).round(2)
by_airline[:10]
```

The code in column 0 corresponds to IATA's abbreviaton for airlines name. Once I've converted the abbreviation, the result indicates that the three airlines with more operating flights are: **SouthWest Airlines, American Airlines, and Envoy Airlines**

```python
#column diverted
#How many flights have been diverted?

total_diverted=sum(airlines.Diverted)
print('There has been',total_diverted,'flights diverted')
percentage_diverted_flights=round((total_diverted/total_flights_year)*100,3)
print("The % of diverted flights over total flights is", percentage_diverted_flights,'%')
```

```python
#Check how many flights have a delay on arrival
#First, I will add a column with the condition

subset_airlines['DelayOnArrival']=(subset_airlines['ArrDelay']>0.0)
subset_airlines.head()
```

```python
#Now, will count all the values representing a delay on arrival

true_arrival_delays=subset_airlines.DelayOnArrival.sum()
print('There are',true_arrival_delays,'flights delayed on arrival')
```

```python
#In terms of % over total flights

percent_arrDelay=((true_arrival_delays/by_month['Number of flights'].sum())*100).round(2)
percent_arrDelay
```

So, 88.98% of flights have experienced a delay on arrival...

#### Taula de les aerolínies amb més endarreriments acumulats

```python
#we already have a dataset with the number of flights per year by airline and we also have the sum of all delays upon arrival
#we need to calculate the number of delays upon departure

subset_airlines['DepDelay']=(subset_airlines['DepDelay']>0.0)
true_dep_delays=subset_airlines.DepDelay.sum()
percent_depart_delay=((true_dep_delays/by_month['Number of flights'].sum())*100).round(2)
print('There are',true_dep_delays,'flights delayed on departure per year which represent',percent_depart_delay,'% of annual flights')
```

Basically, it means that all flights in the dataset where delayed upon departure.

```python
#Now, let's see what airlines are the most affected by delays. As 100% of flights are delayed on departure we will
#group the results by delayed on arrival
#let's use groupby to group all the variables we need by airline. This means “Group the rows by UniqueCarrier and then 
#add up all the values with the same UniqueCarrier

grouped=subset_airlines.groupby('UniqueCarrier')
delays_on_arrival=(grouped['DelayOnArrival'].agg(np.sum)).to_frame()
delays_on_arrival['Nr. flights per year']=airlines.UniqueCarrier.value_counts()
delays_on_arrival['% of delayed flights on arrival']=((delays_on_arrival['DelayOnArrival']/delays_on_arrival['Nr. flights per year'])*100).round(2)
delays_on_arrival.sort_values(['% of delayed flights on arrival'],inplace=True,ascending=False)
delays_on_arrival[:10]
```

#### Quins són els vols més llargs? I els més endarrerits?

```python
#vols més llargs: ordenarem les dades per distància entre origen i destinació
#fem servir groupby i la funció de numpy average per trobar la distància mitjana de cadascuna de les observacions per
#aerolínia, origen i destinació

grouped=subset_airlines.groupby(['Origin','Dest'])
longest_flights=(grouped['Distance']).agg(np.average).to_frame()
longest_flights.sort_values(by=['Distance'],ascending=False)
```

So, the 3 longest flights by distance in the database are:

**Honolulu-Newark** and viceversa

**Honolulu-Atlanta** and viceversa

**Chicago-Honolulu** and viceversa

I've checked the distances on Google and the result is consistent.

```python
#the most delayed flights
#I will add the airlines in this case and take the average arrival delayed time per origin and destination

grouped=subset_airlines.groupby(['UniqueCarrier','Origin','Dest'])
most_delayed=(grouped['ArrDelay']).agg(np.average).round(1).to_frame()
most_delayed=most_delayed.sort_values(by=['ArrDelay'],ascending=False)
most_delayed[:10]
```

The following flights are the ones which accumulates more delays on average:

**Champaign, IL (CMI) to Springfield, IL (SPI)**

**Washington D.C (DCA) to Indianapolis, IND (IND)**

**Cedar Rapids,IA (CID) to Denver, CO (DEN)**
